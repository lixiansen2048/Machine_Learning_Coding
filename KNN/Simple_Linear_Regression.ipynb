### 实现 Simple Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5],dtype=float)
y = np.array([1,3,2,3,5],dtype=float)
```

```python
plt.scatter(x,y)
plt.axis([0,6,0,6])
plt.show()
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
```

```python
num = 0.0
d = 0.0
for x_i,y_i in zip(x,y):
    num += (x_i - x_mean) * (y_i - x_mean)
    d += (x_i - x_mean) ** 2
a = num / d
b = y_mean - a * x_mean
```

```python
a
```

```python
b
```

```python
y_hat = a * x + b
```

```python
plt.plot(x,y_hat,color='r')
plt.scatter(x,y)
plt.axis([0,6,0,6])
plt.show()
```

```python
x_predict = 6
y_predict = a * x_predict + b
y_predict
```
