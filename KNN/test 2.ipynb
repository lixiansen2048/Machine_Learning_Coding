```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
```

```python
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
```

```python
cosy = np.cos(x)
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy)
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color = 'red',linestyle = '--')
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color = 'red',linestyle = '--')
plt.xlim(-5,15)
plt.ylim(0,1.5)
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color = 'red',linestyle = '--')
plt.axis([-1,11,-2,2])
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color = 'red',linestyle = '--')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
plt.plot(x,siny,label = 'sin(x)')
plt.plot(x,cosy,color = 'red',linestyle = '--',label = 'cos(x)')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

```python
plt.plot(x,siny,label = 'sin(x)')
plt.plot(x,cosy,color = 'red',linestyle = '--',label = 'cos(x)')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.title("welcome to the ML world")

plt.show()
```
