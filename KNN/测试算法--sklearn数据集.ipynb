## train_test_split(训练集与测试集分离)

方法一：将 X 与 y 合并后乱序再分离

```python
Z = np.hstack([X,y.reshape(-1,1)])
```

```python
np.random.shuffle(Z)
```

```python
X1,y1 = np.split(Z,[4],axis=1)
```

```python
test_ratio = 0.2
test_num = int(len(X1) * test_ratio)
test_num
```

```python
X_test = X1[:test_num]
X_train = X1[test_num:]
y_test = y1[:test_num]
y_train = y1[test_num:]
```

方法二：将索引乱序后分离

```python
shuffle_indexes = np.random.permutation(len(X))
```

```python
shuffle_indexes
```

```python
test_ratio = 0.2
test_size = int(len(X)*test_ratio)
```

```python
test_size
```

```python
test_indexes = shuffle_indexes[:test_size]
train_indexes = shuffle_indexes[test_size:]
```

```python
test_indexes
```

```python
X_train = X[train_indexes]
y_train = y[train_indexes]
print(X_train.shape)
```
