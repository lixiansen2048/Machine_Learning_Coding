### 从高维数据向低维数据的映射

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.empty((100,2))
```

```python
X.shape
```

```python
X[:,0] = np.random.uniform(0.,100.,size=100)
X[:,1] = 0.75 * X[:,0] + 3 + np.random.normal(0.,10.,size=100)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
from playML.PCA import PCA
```

```python
pca = PCA(n_component=2)
```

```python
pca.fit(X)
```

```python
pca.component_
```

```python
pca = PCA(n_component=1)
```

```python
pca.fit(X)
```

```python
X_reduction = pca.transform(X)
```

```python
X_reduction.shape
```

```python
X_restore = pca.inver_trainsform(X_reduction)
```

```python
X_restore.shape
```

```python
plt.scatter(X[:,0],X[:,1],color='r',alpha=0.5)
plt.scatter(X_restore[:,0],X_restore[:,1],color='b',alpha=0.5)
plt.show()
```

### scikit-learn 中封装的 PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=1)
```

```python
pca.fit(X)
```

```python
X_reduction = pca.transform(X)
```

```python
X_reduction.shape
```

```python
X_restore = pca.inverse_transform(X_reduction)
```

```python
X_restore.shape
```

```python
plt.scatter(X[:,0],X[:,1],color='r',alpha=0.5)
plt.scatter(X_restore[:,0],X_restore[:,1],color='b',alpha=0.5)
plt.show()
```
